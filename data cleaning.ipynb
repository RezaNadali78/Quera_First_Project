## unit_per_floor

```python
def convert_unit(x):
    if x == 'more_than_8':
        return 9
    elif x == 'unselect':
        return float('nan')
    return float(x)
data.unit_per_floor = data.unit_per_floor.apply(lambda x: convert_unit(x))
```

## total_floors_count

```python
def convert_total(x):
    if x == '30+':
        return 31
    elif x == 'unselect':
        return float('nan')
    return float(x)
data.total_floors_count = data.total_floors_count.apply(lambda x: convert_total(x))
```

## rooms_count

```python
def transform(u):
    if u == 'بدون اتاق':
        return int(0)
    elif u == 'یک':
        return int(1)
    elif u == 'دو':
        return int(2)
    elif u == 'سه':
        return int(3)
    elif u == 'چهار':
        return int(4)
    elif u == 'پنج یا بیشتر':
        return int(5)
```

```python
data.rooms_count = data.rooms_count.apply(lambda x: transform(x))
data.rooms_count
```

## created_at_month

```python
data['month'] = pd.to_datetime(data['created_at_month']).dt.month
data['year'] = pd.to_datetime(data['created_at_month']).dt.year
data = data.drop('created_at_month',axis=1)
data
```

## construcion_year

```python
def convert_construction_year(x):
    if x == 'قبل از ۱۳۷۰':
        return 1370
    elif x == float('nan'):
        return x
    else:
        x = str(x)
        persian_digits ='۱۲۳۴۵۶۷۸۹۰١٢٣٤٥٦٧٨٩٠'
        english_digits ='12345678901234567890'
        translation_table = str.maketrans(persian_digits, english_digits)
        x = float(x.translate(translation_table))
        return x
```

```python
data['construction_year'] = data.construction_year.apply(convert_construction_year).astype('Int64')
```

# floor

```python
data.floor.unique()
```

```python
def convert_floor(x):
    if x == '30+':
        return 30
    return float(x)
data.floor = data.floor.apply(lambda x: convert_floor(x))
```

```python
df = data.copy()
```

```python
df = df[~(df['price_value'].isna()) | ~(df['transformed_credit'].isna())]
df = df[~(df['location_longitude'].isna()) & ~(df['location_longitude'].isna())]
```

```python
object_cols = df.select_dtypes(include=['object']).columns
object_cols
```

## Encoding Object Features

```python
df = pd.get_dummies(df, columns = object_cols,dtype = int, prefix_sep='_')
df
```

# Feature Selection

```python
from sklearn.model_selection import train_test_split
target = ['price_value','transformed_credit']
X_train, X_test, y_train, y_test = train_test_split(
    df.drop(target, axis=1), 
    df[target],
    test_size=0.3,
    random_state=42
)
```

```python
xy_train_price = pd.concat([X_train,y_train.price_value],axis=1)
xy_train_credit = pd.concat([X_train,y_train.transformed_credit],axis=1)
xy_test_price = pd.concat([X_test,y_test.price_value],axis=1)
xy_test_credit = pd.concat([X_test,y_test.transformed_credit],axis=1)
```

## corrlation between features and price_value by spearman method

```python
corr_price = pd.DataFrame(xy_train_price.corr(method='spearman').price_value)
```

```python
feature_of_price = corr_price[(corr_price.price_value > 0.25) | (corr_price.price_value < -0.25)].index
```

## corrlation between features and transformed_credit by spearman method

```python
corr_credit = pd.DataFrame(xy_train_credit.corr(method='spearman').transformed_credit)
```

```python
feature_of_credit = corr_credit[(corr_credit.transformed_credit > 0.25) | (corr_credit.transformed_credit < -0.25)].index
```

# Featuring of Train And Test

```python
xy_train_price = xy_train_price[feature_of_price]
xy_test_price = xy_test_price[feature_of_price]
xy_train_credit = xy_train_credit[feature_of_credit]
xy_test_credit = xy_test_credit[feature_of_credit]
```

```python
xy_train_price = xy_train_price[~xy_train_price.price_value.isna()]
xy_test_price = xy_test_price[~xy_test_price.price_value.isna()]
xy_train_credit = xy_train_credit[~xy_train_credit.transformed_credit.isna()]
xy_test_credit = xy_test_credit[~xy_test_credit.transformed_credit.isna()]
```
